```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def sigmoid(x):
    sig = 1/(1+np.exp(-x))
    return sig 
```

```python
x = np.linspace(-10,10,100)
y = sigmoid(x)
plt.plot(x,y)
```

```python
#softmax
data = [1,8,3,5]
print(np.max(data))
print(np.argmax(data))
```

```python
def softmax(vector):
    e = np.exp(vector)
    return e/e.sum()
```

```python
x = np.linspace(-10,10,100)
y = softmax(x)
plt.plot(x,y)
```
